```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataframe = pd.read_csv('bank_transactions_data_2 1.csv')
```

```python
dataframe
```

```python
dataframe.head(10)
```

```python
dataframe.info()
```

```python
plt.plot(dataframe["TransactionID"], dataframe['TransactionAmount'] )
```

```python
plt.plot(dataframe["TransactionID"], dataframe['TransactionAmount'] )
plt.title("Bank Transactions")
plt.xlabel("TransactionID")
plt.ylabel("Transactions Amount")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(dataframe['TransactionAmount'], dataframe['AccountBalance'], alpha=0.6)
plt.title('Scatter Plot: Account Balance vs Transaction Amount', fontsize=14)
plt.xlabel('Transaction Amount', fontsize=12)
plt.ylabel('Account Balance', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(dataframe['TransactionAmount'], dataframe['AccountBalance'], alpha=0.6)
plt.title('Scatter Plot: Account Balance vs Transaction Amount', fontsize=14)
plt.xlabel('Transaction Amount', fontsize=12)
plt.ylabel('Account Balance', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(dataframe['CustomerAge'], bins=20, color='green', alpha=0.7, edgecolor='black')
plt.title('Histogram of Customer Age', fontsize=14)
plt.xlabel('Customer Age', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(data=dataframe, x='TransactionAmount', fill=True, color='blue')
plt.title('Kernel Density Plot of Transaction Amount', fontsize=14)
plt.xlabel('Transaction Amount', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
